# Fruit ninja

```python
import pandas as pd

# Create a dataset of fruits
fruits_df = pd.DataFrame({
    'Fruit': ['Apple', 'Banana', 'Orange', 'Strawberry', 'Mango', 'Grapes', 'Watermelon', 'Blueberry'],
    'Color': ['Red', 'Yellow', 'Orange', 'Red', 'Yellow', 'Purple', 'Green', 'Blue'],
    'Price': [0.50, 0.30, 0.60, 1.00, 0.80, 1.50, 2.00, 2.50],
    'Quantity': [50, 100, 80, 40, 60, 70, 30, 25]
})

print(fruits_df)
```

```python
import matplotlib.pyplot as plt

plt.scatter(fruits_df['Price'],fruits_df['Fruit'])
plt.xlabel('Price')
plt.ylabel('Fruit')
plt.title('Price vs Fruit')
plt.show()
```
